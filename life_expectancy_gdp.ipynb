```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

all_data = pd.read_csv('all_data.csv')
#print(all_data.head())
print(np.mean(all_data['Life expectancy at birth (years)']))
print(np.median(all_data['Life expectancy at birth (years)']))
```

The mean overall life expectancy for all these nations is 72.78

The median overall life expectancy for these nations is 76.75

```python
# create a list of nation names
nations = ['Chile', 'China', 'Mexico', 'United States of America', 'Zimbabwe']
# create a dictionary 
nation_results = {}

for nation in nations:
    all_data = pd.read_csv('all_data.csv')
    
    
    all_data = all_data[all_data['Country'] == nation]
    
    
    # Calculate the mean and median life expectancy and GDP for the nation
    mean_le = all_data['Life expectancy at birth (years)'].mean()
    median_le = all_data['Life expectancy at birth (years)'].median()
    mean_gdp = all_data['GDP'].mean()
    median_gdp = all_data['GDP'].median()
    
    
    # Store the results in the dictionary
    nation_results[nation] = {
        'mean_life_expectancy': mean_le,
        'median_life_expectancy': median_le,
        'mean_GDP': mean_gdp,
        'median_GDP': median_gdp
    }
        
        # Print out the results
for nation, data in nation_results.items():
    print(f"{nation}:")
    print(f"Mean life expectancy: {data['mean_life_expectancy']}")
    print(f"Median life expectancy: {data['median_life_expectancy']}")
    print(f"Mean GDP: {data['mean_GDP']}")
    print(f"Median GDP: {data['median_GDP']}\n")
```

As we see the results for each nation the mean and median life expectancy for Zimbabwe is way below the total average and also much lower then the averages of each of the individual nations. The GDP is also much lower then the others and that be contributing to the lower life expectancy. 
Unexpectedly though Chile has the highest mean and median life expectancy out of these nations beating out the US even though Chile has a lower GDP then the US. So GDP might not be the best way of determining the life expectancy of a nation.
Now we will plot this imformation into graphs to make it easier to understand and read.


```python
for nation in nations:
    # Read the CSV file into a pandas dataframe
    all_data = pd.read_csv('all_data.csv')

    # Filter the dataframe to only include rows for the current nation
    all_data = all_data[all_data['Country'] == nation]

    # Create a line plot of life expectancy over time
    plt.plot(all_data['Year'], all_data['Life expectancy at birth (years)'], label=nation)

# Set the x- and y-axis labels
plt.xlabel('Year')
plt.ylabel('Life Expectancy')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
plt.clf()
```

This line graph shows the large difference in life expectancy between Zimbabwe and the other nations. There also seems to have been a dip in life expectancy in Zimbabwe that is not present in the other nations. Might have been some internal conflict within Zimbabwe at the time such as war or famine or disease that caused the life expectancy to go down. But since then it has grown alot. Where with the other nations you can see Chile has been in the lead the whole time and growing steadily as well.

When plotting all the nations on a single scatter plot it is hard to see if the GDP effects the life expectancy so lets plot them individually and see what that looks like.

```python
for nation in nations:
    all_data = pd.read_csv('all_data.csv')
   

    # Filter the dataframe to only include rows for the current nation
    all_data = all_data[all_data['Country'] == nation]

    # Create a scatter plot of GDP versus life expectancy for the current nation
    plt.scatter(all_data['GDP'], all_data['Life expectancy at birth (years)'], label=nation)

    # Set the x- and y-axis labels
    plt.xlabel('GDP')
    plt.ylabel('Life Expectancy')

    # Add a title to the plot
    plt.title(f"GDP versus Life Expectancy for {nation}")

    # Add a legend to the plot
    plt.legend()

    # Show the plot
    plt.show()
    plt.clf()
```

When plotted individually we can now see that the rise of GDP increases the life expectancy for all of these nations. In the case of the drip in life expectancy for Zimbabwe we can see that their GDP also dropped so the lowest level of life expectancy and lowest GDP occured at the same time.

Earlier we made the assumption that when the GDP goes down so does the life expactancy, but that is not entirely true. As we saw with Zimbabwe the life expectancy was lowest in 2004 but their GDP was at its lowest in 2008 when the life expectancy was already starting to rise. Same with the US which had a GDP drop in 2009 but life expectancy did not drop at that time. Mexico on the other hand had a drop in GDp in 2009 and life expectancy dropped in 2008 and rose a little in 2009 and again dropped in 2010. China on the other hand has had its GDP steadly rise and its life expectancy also rose at the same time. With Chile its GDP too a drop in 2014-2015 while its life expectancy contiunes to grow. So GDP and life expectancy do have some corrilation, i.e. if GDP goes up so does life expectancy but life expectancy doesnt always go down when the GDP goes down.
